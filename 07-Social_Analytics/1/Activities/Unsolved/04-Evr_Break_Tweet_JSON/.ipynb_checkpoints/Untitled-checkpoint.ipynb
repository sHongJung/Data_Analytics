```python
# Dependencies
import tweepy
import json
import twitter_api as ta
from pprint import pprint

consumer_key = ta.consumer_key
consumer_secret = ta.consumer_secret
access_token = ta.access_token
access_token_secret = ta.access_token_secret 

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())

# Read JSONs
sample = "Sample.json"
```

```python
with open(sample) as data_file:
    sample_data = json.load(data_file)
    
    pprint(sample_data)
```

```python
 # Question 1: With what account are the tweets in the Sample associated?
print(sample_data[0]['user']['name'])
```

```python
# Question 2: How many followers does this account have associated with it?
print(sample_data[0]['user']['followers_count'])
```

```python
# Question 3: How many tweets are included in the Sample? 
count = 0
for tweet in sample_data:
    count = count + 1
print(count)
```

```python
# Question 4: How many tweets total has this account made?
print(sample_data[0]['user']['statuses_count'])
```

```python
# Question 5: What was the text in the most recent tweet?
# Hint: You may need to look into "encding utf-8 in python"
print(sample_data[0]['text'])
```

```python
# Question 6: What was the text associated with each of the tweets
# included in this sample data?
for tweet in sample_data:
    print(sample_data[0]['text'])
```

```python
# Question 7 (Bonus): Which of the user's tweets was most frequently
# retweeted? How many retweets were there?
sorted_data = sorted(sample_data, reverse=True, key=lambda k:k['retweet_count'])
print(f"Most frequently retweeted {sorted_data[0]['text']}\nCount: {sorted_data[0]['retweet_count']}")
```
