## Variance, standard deviation and z score

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
arr1 = np.array([3, 4, 5, 6, 7])
arr2 = np.array([-1525, -200, 5, 745, 1000])

# Calculate mean, median and mode
print('Mean 1: ', np.mean(arr1))
print('Mean 2: ', np.mean(arr2))

print('Median 1: ', np.median(arr1))
print('Median 2: ', np.median(arr1))
```

```python
xrange = (-1, 5)
lim = 1.1*max(abs(arr2))
yrange = (-lim, lim)

# Create figure
fig = plt.figure(figsize=(20, 5))

# Make plot on the left
fig.add_subplot(1, 2, 1)
plt.plot(arr1, 'o', markersize=10)
plt.hlines(np.mean(arr1), xrange[0], xrange[1], 'r')
plt.xlim(xrange)
plt.ylim(yrange)

# Make plot on the left
fig.add_subplot(1, 2, 2)
plt.plot(arr2, 'o', markersize=10)
plt.hlines(np.mean(arr2), xrange[0], xrange[1], 'r')
plt.xlim(xrange)
plt.ylim(yrange)

plt.show()
```

```python
# Calculate variance and standard deviation
print('Variance 1: ', np.var(arr1))
print('Variance 2: ', np.var(arr2))

print('Standard Deviation 1: ', np.std(arr1))
print('Standard Deviation 2: ', np.std(arr2))
```

```python
# Create figure
fig = plt.figure(figsize=(20, 5))

# Make plot on the left
fig.add_subplot(1, 2, 1)
plt.plot(arr1, 'o', markersize=10)
plt.hlines(np.mean(arr1), xrange[0], xrange[1], 'r')
lower = np.mean(arr1)-np.std(arr1)
upper = np.mean(arr1)+np.std(arr1)
plt.hlines([lower, upper], xrange[0], xrange[1], 'g',
           linestyle='dotted')
plt.xlim(xrange)

# Make plot on the left
fig.add_subplot(1, 2, 2)
plt.plot(arr2, 'o', markersize=10)
plt.hlines(np.mean(arr2), xrange[0], xrange[1], 'r')
lower = np.mean(arr2)-np.std(arr2)
upper = np.mean(arr2)+np.std(arr2)
plt.hlines([lower, upper], xrange[0], xrange[1], 'g',
           linestyle='dotted')
plt.xlim(xrange)

plt.show()
```

```python
arr = np.array([30, 31, 31, 32, 32, 40, 41, 41, 1000])
avg = np.mean(arr)
stdev = np.std(arr)
print('Mean: ', avg)
print('Standard Deviation: ', stdev)
```

```python
xrange = (-1, len(arr))
plt.xlim(xrange)
plt.plot(arr, 'o', markersize=10)
plt.hlines(avg, xrange[0], xrange[1], 'r')
lower = avg-stdev
upper = avg+stdev
plt.hlines([lower, upper], xrange[0], xrange[1], 'g',
           linestyle='dotted')
plt.title(arr.__repr__())
plt.show()
```

```python
# Z score of a given point is how far away is the point from the mean,
# measured in units of standard deviations
stats.zscore(arr)
```
