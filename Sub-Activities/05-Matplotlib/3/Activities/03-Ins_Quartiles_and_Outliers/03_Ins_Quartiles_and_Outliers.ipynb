## Quartiles and outliers

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Extreme values in a dataset are called `outliers`
#    - Outliers can skew your descriptive statistics
#    - Can I ignore them?
arr = np.array([30, 31, 31, 32, 32, 40, 41, 41, 1000])
plt.plot(arr, 'o', markersize=10)
plt.ylabel('Height of my koala bears (in inches)')
plt.show()
```

```python
npts = 200
x_values = np.random.randn(npts)
x_values = np.append(x_values, 13.3)
y_values = np.random.randn(npts)
y_values = np.append(y_values, 12.7)
fig, ax = plt.subplots()
ax.scatter(x_values, y_values)
ax.set_aspect('equal')

# Linear regression
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xarr = np.arange(xlim[0], xlim[1])
slope, intercept, _, _, _ = stats.linregress(x_values, y_values)
line = intercept + slope * xarr
ax.plot(xarr, line, 'r--')
ax.set_xlim(xlim)
ax.set_ylim(ylim)
plt.show()
```

```python
npts = 200
x_values = np.random.randn(npts)
x_values = np.append(x_values, 8.3)
y_values = np.random.randn(npts+1)*0.5 + 0.7*x_values

fig, ax = plt.subplots()
ax.scatter(x_values, y_values)
ax.set_aspect('equal')

# Linear regression
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xarr = np.arange(xlim[0], xlim[1])
slope, intercept, _, _, _ = stats.linregress(x_values, y_values)
line = intercept + slope * xarr
ax.plot(xarr, line, 'r--')
ax.set_xlim(xlim)
ax.set_ylim(ylim)
plt.show()
```

```python
npts = 200
x_values = np.random.randn(npts)
y_values = np.random.randn(npts)*0.7 + 0.7*x_values
x_values = np.append(x_values, 19.3)
y_values = np.append(y_values, 4.3)


fig, ax = plt.subplots()
ax.scatter(x_values, y_values)

# Linear regression
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xarr = np.arange(xlim[0], xlim[1])
slope, intercept, _, _, _ = stats.linregress(x_values, y_values)
line = intercept + slope * xarr
ax.plot(xarr, line, 'r--')
ax.set_xlim(xlim)
ax.set_aspect('equal')

plt.show()
```

One way to remove outliers is to remove the points that are "far away" from the median.


```python
npts = 1001
arr = np.random.randn(npts)

# Calculate median
med = np.median(arr)
quartiles = np.percentile(arr, [25, 75])
```

```python
lower_quartile, upper_quartile = np.percentile(arr, [25, 75])
print('Quartiles: ', lower_quartile, ',', upper_quartile)

interquartile_range = upper_quartile-lower_quartile
print('Interquartile range: ', interquartile_range)
```

```python
xrange = (0, npts-1)
lim = 1.25*max(abs(arr))
yrange = (-lim, lim)

# Create figure
fig = plt.figure(figsize=(20, 5))

# Make plot on the left
fig.add_subplot(1, 2, 1)
plt.plot(arr, '.')
plt.xlim(xrange)
plt.ylim(yrange)

# Make plot on the right
fig.add_subplot(1, 2, 2)
plt.plot(np.sort(arr), '.')
plt.hlines(med, xrange[0], xrange[1], 'r')
plt.vlines((npts-1)/2, yrange[0], yrange[1],
           linestyle='dotted')
plt.hlines(quartiles, xrange[0], xrange[1], 'g')
plt.vlines([(npts-1)/4, 3*(npts-1)/4], yrange[0], yrange[1],
           linestyle='dotted')

plt.xlim(xrange)
plt.ylim(yrange)

plt.show()
```
