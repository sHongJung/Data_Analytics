```python
# Dependencies
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np


x_values = np.arange(10)
fake = [1, 2.5, 2.75, 4.25, 5.5, 6, 7.25, 8, 8.75, 9.8]

slope, intercept, _, _, _ = stats.linregress(x_values, fake)
fit = slope * x_values + intercept

fig, ax = plt.subplots()

ax.plot(x_values, fake, linewidth=0, marker='o')
ax.plot(x_values, fit, 'b--')

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

ax.set_xlabel("Fake Banana Ages (in days)")
ax.set_ylabel("Fake Banana Weights (in Hundreds of Kilograms)")

plt.show()
```
