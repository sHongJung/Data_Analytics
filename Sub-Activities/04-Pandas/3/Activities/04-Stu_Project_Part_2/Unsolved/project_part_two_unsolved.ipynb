```python
# Dependencies
import pandas as pd
import os
```

```python
# load CSV
csv_path = os.path.join('..', 'Resources', '2016-FCC-New-Coders-Survey-Data.csv')
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
bootcamp_df = pd.read_csv(csv_path)
bootcamp_df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
bootcamp_df = bootcamp_df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
bootcamp_df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
bootcamp_df = bootcamp_df.replace([0.0, 1.0], ["No", "Yes"])
bootcamp_df.head()
```

```python
# Extract rows for only those who attended a bootcamp
bootcamp_df = bootcamp_df.loc[bootcamp_df["AttendedBootcamp"] == "Yes"]
bootcamp_df.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
grouped_bootcamp = bootcamp_df.groupby('BootcampName')
grouped_bootcamp["BootcampName"].value_couts()
```
