```python
# Dependencies
import pandas as pd
import os
```

```python
# load 2016-FCC-New-Coders-Survey-Data.csv
csv_path = os.path.join('..', 'Resources', '2016-FCC-New-Coders-Survey-Data.csv')
```

```python
# Read with pandas
bootcamp_df = pd.read_csv(csv_path)
bootcamp_df.head()
```

```python
# Inspect all columns
bootcamp_df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111 
bootcamp_df1 = bootcamp_df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
bootcamp_df1.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
def changeResponse(value):
    response = ""
    if value == 0.0:
        response = "No"
    elif value == 1.0:
        response = "Yes"
    return response

bootcamp_df1["AttendedBootcamp"] = bootcamp_df1["AttendedBootcamp"].map(changeResponse)
bootcamp_df1.head()
```

```python
# Calculate total number of respondents in survey
total_surveyed = len(bootcamp_df1)
total_surveyed
```
