```python
import pandas as pd
import os
```

```python
bitcoin_path = os.path.join('Resources', 'bitcoin_cash_price.csv')
dash_path = os.path.join('Resources', 'dash_price.csv')

bitcoin_df = pd.read_csv(bitcoin_path)
dash_df = pd.read_csv(dash_path)

bitcoin_df.head(10)
```

```python
dash_df.head(10)
```

```python
merge_table = pd.merge(bitcoin_df, dash_df, on="Date", how = "outer")
merge_table.head()
```

```python
merge_table.rename(columns={"Open_x" : "Bitcoin Open",
                            "Close_x" : "Bitcoin Close",
                            "Open_y" : "Dash Open",
                            "Close_y" : "Dash Close" })
    
```
