```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
bk_df = pd.read_csv("Resources\got.csv")
bk_df.head()
```

```python
# Get attacker and defender data
bk_df = bk_df[["attacker_king", "defender_king"]]
bk_df.head(10)
```

```python
# Get total battle data
attack_battle = bk_df["attacker_king"].value_counts()

defence_battle = bk_df["defender_king"].value_counts()

total_battle = attack_battle.add(defence_battle, fill_value=0)
total_battle.head()
```

```python
# Configure plot and ticks
total_battle.plot(kind="bar", color="red")
```
