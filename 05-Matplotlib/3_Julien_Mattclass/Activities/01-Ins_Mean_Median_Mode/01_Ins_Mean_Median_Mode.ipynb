## Mean, median and mode

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import random
```

```python
# Define a numpy array
arr = np.array([6, 1, 5, 6, 2])
```

```python
# What is the mean of this array?
avg = np.mean(arr)

# What is the median of this array?
med = np.median(arr)

# What is the mode of this array?
mode = stats.mode(arr)
mode = mode.mode[0]
```

```python
# Define line colors and styles
colors = ['steelblue', 'tomato', 'limegreen']
styles = ['dashed', 'dotted', 'solid']
random.shuffle(colors)
random.shuffle(styles)

# Define x-axis limits
xrange = (-1, 5)
plt.xlim(xrange)

# Create plot and add horizontal lines
plt.plot(arr, 'o', markersize=10)
plt.hlines([avg, med, mode], xrange[0], xrange[1],
           colors=colors, linewidth=2, linestyles=styles)
plt.title(arr.__repr__())
plt.show()
```

```python
xrange = (-1, 5)
plt.xlim(xrange)

# Create plot and add horizontal lines
plt.plot([2, 1, 2, 3, 2], 'o', markersize=10)
plt.hlines(2, xrange[0], xrange[1], 'r')
plt.show()
```

```python
npts = 1001
arr = np.random.randn(npts)

# Calculate mean, median and mode
avg = np.mean(arr)
med = np.median(arr)
mode = stats.mode(arr)
mode = mode.mode[0]
print('Mean: ', avg)
print('Median: ', med)
print('Mode: ', mode)
```

```python
xrange = (0, npts-1)
lim = 1.25*max(abs(arr))
yrange = (-lim, lim)

# Create figure
fig = plt.figure(figsize=(20, 5))

# Make plot on the left
fig.add_subplot(1, 2, 1)
plt.plot(arr, '.')
plt.hlines(avg, xrange[0], xrange[1], 'r')
plt.xlim(xrange)
plt.ylim(yrange)

# Make plot on the right
fig.add_subplot(1, 2, 2)
plt.plot(np.sort(arr), '.')
plt.hlines(med, xrange[0], xrange[1], 'g')
plt.vlines((npts-1)/2, yrange[0], yrange[1],
           linestyle='dotted')
plt.xlim(xrange)
plt.ylim(yrange)

plt.show()
```

### What's the best number to describe your data?

```python
arr = np.array([30, 31, 31, 32, 32, 40, 41, 41, 1000])
avg = np.mean(arr)
med = np.median(arr)
```

```python
xrange = (-1, len(arr))
plt.xlim(xrange)
plt.plot(arr, 'o', markersize=10)
plt.hlines([avg, med], xrange[0], xrange[1],
           colors=colors, linewidth=2, linestyles=styles)
plt.title(arr.__repr__())
plt.show()
```

```python
arr = np.array([30, 31, 31, 32, 32, 40, 41, 41, 1000,
                1210, 1210, 1567])
avg = np.mean(arr)
med = np.median(arr)
```

```python
xrange = (-1, len(arr))
plt.xlim(xrange)
plt.plot(arr, 'o', markersize=10)
plt.hlines([avg, med], xrange[0], xrange[1],
           colors=colors, linewidth=2, linestyles=styles)
plt.title(arr.__repr__())
plt.show()
```

##### What about the mode?

```python
# Can there be several modes? YES

modes = [31, 32, 41, 1210] 
# However, scipy.stats.mode will
# only return the smallest value in this list

plt.xlim(xrange)
plt.plot(arr, 'o', markersize=10)
plt.hlines(modes, xrange[0], xrange[1], color='r', linestyle=':')
plt.title(arr.__repr__())
plt.show()
```
